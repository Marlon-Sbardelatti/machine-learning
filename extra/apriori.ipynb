**1. Acompanhe o vídeo abaixo para implementar o algoritmo de Regras de Associação do Mercado.**

```python
import pandas as pd
from apyori import apriori

market_data = pd.read_csv('mercado.csv', header = None)

transactions = []
for i in range(0, 10):
  transactions.append([str(market_data.values[i, j]) for j in range(0, 4)])

rules = apriori(transactions, min_support=0.3, min_confidence=0.8, min_lift=2, min_length=2)

results = list(rules)
results
```

**a) Vamos supor que você quer extrair Regras da Associação para os produtos que são vendidos pelo menos quatro vezes ao dia, ou seja, 28 vezes por semana (4*7). Adicione o valor de suporte de acordo com o problema. Execute o algoritmo e responda qual é o resultado.**

```python
transactions = []
for i in range(0, 10):
  transactions.append([str(market_data_2.values[i, j]) for j in range(0, 4)])

support = 28 / len(market_data_2.values)

rules = apriori(transactions, min_support = support)

results = list(rules)
results
```

**b) Adicione o valor de confiança para 0.2 e o lift para 3. Quantos registros foram retornados?**

```python
rules = apriori(transactions, min_support=support, min_confidence=0.2, min_lift=3)

results = list(rules)
print('Número de registros:', len(results))
```

**c) Adicione o código abaixo para poder visualizar melhor os dados. Faça comentário do que acontece em cada linha do código.**

```python
# Cria vetores para armazenar itens dos itemsets
A = []
B = []

# Cria vetores para armazenar métricas de cada itemset
suporte = []
confianca = []
lift = []

# Itera sobre os resultados calculados
for resultado in results:

  # Recupera suporte
  s = resultado[1]

  # Recupera regras para itemset
  result_rules = resultado[2]

  for result_rule in result_rules:
    # Recupera itens do itemset (ex: a = almonds e b = avocado)
    a = list(result_rule[0]) # SE
    b = list(result_rule[1]) # ENTÃO

    # Recupera confiança para a regra
    c = result_rule[2]

    # Recupera lift para a regra
    l = result_rule[3]

    # Adiciona itens do itemset no respectivo vetor
    A.append(a) # a -> Lista de "SE"
    B.append(b) # b -> Lista de "ENTÃO"

    # Adiciona suporte da regra na lista de suportes
    suporte.append(s)

    # Adiciona confiança da regra na lista de confianças
    confianca.append(c)

    # Adiciona lift da regra na lista de lifts
    lift.append(l)

'''
  Cria um dataframe a partir das listas criadas, onde, para cada índice, é possível
  produzir uma linha com todas as métricas da regra, que se tornam colunas:
  - A: Item "SE"
  - B: Item "ENTÃO"
  - suporte: Suporte da regra
  - confianca: Confiança da regra
  - lift: Lift da regra

  Exemplo: SE almonds ENTÃO avocado, com suporte de 0.1, confiança de 1.0 e lift de 10.
'''
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})

# Ordena linhas do dataframe (regras) pelo atributo de lift de forma descendente, ou seja, do maior pro menor
rules_df.sort_values(by = 'lift', ascending = False)
```
