**1. Acompanhe o vídeo abaixo para implementar o algoritmo de Regras de Associação do Mercado.**

```python
import pandas as pd
from apyori import apriori

market_data = pd.read_csv('mercado.csv', header = None)

transactions = []
for i in range(0, 10):
  transactions.append([str(market_data.values[i, j]) for j in range(0, 4)])

rules = apriori(transactions)

results = list(rules)
results
```

**2. Utilize a base de dados maior (mercado2) que apresenta o histórico de compras de um supermercado francês em 1 semana. Teste o seu algoritmo e responda as seguintes perguntas:**

```python
market_data_2 = pd.read_csv('mercado2.csv', header = None)
```

**a) Vamos supor que você quer extrair Regras da Associação para os produtos que são vendidos pelo menos quatro vezes ao dia, ou seja, 28 vezes por semana (4*7). Adicione o valor de suporte de acordo com o problema. Execute o algoritmo e responda qual é o resultado.**

```python
transactions = []
for i in range(0, 10):
  transactions.append([str(market_data_2.values[i, j]) for j in range(0, 4)])

support = 28 / len(market_data_2.values)

rules = apriori(transactions, min_support = support)

results = list(rules)
results
```

**b) Adicione o valor de confiança para 0.2 e o lift para 3. Quantos registros foram retornados?**

```python
rules = apriori(transactions, min_support=support, min_confidence=0.2, min_lift=3)

print('Número de registros:', len(list(rules)))
```

**c) Adicione o código abaixo para poder visualizar melhor os dados. Faça comentário do que acontece em cada linha do código.**

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in results:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)

rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
rules_df.sort_values(by = 'lift', ascending = False)
```
